```python
import numpy as np
import matplotlib
import tensorflow as tf
import keras_tuner as kt
from tensorflow import keras
from keras import backend as K
from tensorflow.keras import layers, losses
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2 as cv
from matplotlib.image import imsave, imread
%matplotlib inline
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import tensorflow as tf
from PIL import Image
import pandas as pd
import skimage
from skimage import io as sio
```

```python
data = pd.read_csv("meta-data3.csv")


train_data = np.zeros([5000, 256,256, 4])
train_label=np.zeros([5000,2])
instances =0
counter = 0

for row in data.index:
#Getting the filenames and labels
    if (data['label'][row] == 'NOTHINGHERE'):
            filename = data['file'][row]
            class_ID = 0
            #print(filename, "  ", class_ID, " ", data['sequence'][row], " ", counter)
    else:
            filename = data['file'][row]
            class_ID = 1
            #print(filename, "  ", class_ID, " ", data['sequence'][row], " ", counter)
    
    if counter == 0:
        backSub = cv.createBackgroundSubtractorMOG2()
        init_seq= data['sequence'][row]
        
    cap = cv.imread("Images/" +filename)
    dim =(256,256)
    cap = cv.resize(cap,dim)  
  
    
    #Initialising the subtractor
    if init_seq== data['sequence'][row]:
        fgMask = backSub.apply(cap)
    else:
        counter = 0
        backSub = cv.createBackgroundSubtractorMOG2()
        init_seq= data['sequence'][row]
        fgMask = backSub.apply(cap)
    counter = counter + 1
    
    if counter ==1:
        initial_image = cap
        
    #Saving the file
    if counter ==3:
        counter =0
        seq_no = data['sequence'][row]
        train_ins = np.dstack((initial_image,fgMask))
        #print(train_ins.shape)
        train_data[instances]=train_ins
        if class_ID ==0:
            label_ins = [1,0]
        else:
            label_ins = [0,1]
        train_label[instances]= label_ins
        instances =instances +1
        print("Adding instance ", instances)

actual_train = train_data[0:instances,:,:,:] 
actual_label = train_label[0:instances,:] 
#print(actual_train)
#print(actual_label)
print('train label shape ', actual_train.shape)
print('train data shape ', actual_label.shape)
np.save('train', actual_train)
np.save('label', actual_label)


```

```python
#Loading the numpy arrays-to be used if preprocessing already completed

loaded_train = np.load('train.npy')
loaded_label = np.load('label.npy')
print('after reloading label shape ', loaded_label.shape)
print('after reloading train shape ', loaded_train.shape)
```

```python
#splitting training, testing and validation data

from sklearn.model_selection import train_test_split
# Separate the test data
x, x_test, y, y_test = train_test_split(loaded_train, loaded_label, test_size=0.15, shuffle=True)
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.15, shuffle=True)
```

```python
#Creating CNN model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(256,(3,3),activation = "relu" ,padding='same', input_shape = (256,256,4)) ,
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128,(3,3),padding='same', activation = "relu") ,  
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128,(3,3),padding='same', activation = "relu") ,  
    tf.keras.layers.MaxPooling2D(2,2),
     tf.keras.layers.Conv2D(128,(3,3),padding='same', activation = "relu") ,  
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dense(128,activation ="relu"),
    tf.keras.layers.Dense(64,activation ="relu"),
    tf.keras.layers.Dense(2,activation = "softmax")   #Adding the Output Layer
])
```

```python
# The model looks like this
model.summary()
```

```python
#Compiling the model
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#Training and validating the model
history = model.fit(x=x_train, y=y_train,
                    validation_data=(x_val, y_val),
                    batch_size=30,
                    epochs=5, verbose =2)
```
